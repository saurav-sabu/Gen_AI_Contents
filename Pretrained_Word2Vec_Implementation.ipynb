```python
# Importing necessary modules from the Gensim library
import gensim
from gensim import downloader
```

```python
# Downloading and loading the pre-trained Word2Vec model from Google News dataset
model = downloader.load("word2vec-google-news-300")
```

```python
# Accessing the vector for the word "cricket"
model["cricket"]
```

```python
# Getting the length of the vector for "cricket" which should be 300
len(model["cricket"])
```

```python
# Finding words most similar to "man"
model.most_similar("man")
```

```python
# Finding words most similar to "facebook"
model.most_similar("facebook")
```

```python
# Finding words most similar to "google"
model.most_similar("google")
```

```python
# Calculating the similarity between "facebook" and "social"
model.similarity("facebook","social")
```

```python
# Calculating the similarity between "python" and "animal"
model.similarity("python","animal")
```

```python
# Finding the word that doesn't match in the list ["php", "java", "Python", "c++", "dog"]
model.doesnt_match(["php","java","Python","c++","dog"])
```

```python
# Performing vector arithmetic: vector for "king" minus vector for "man" plus vector for "woman"
vec = model["king"] - model["man"] + model["women"]
vec
```

```python
# Finding words most similar to the resulting vector from the previous step
model.most_similar(vec)
```
